### Lets use a loop to quickly add assemblies into jbrowse

```bash
for fastafiles in *.fasta
  do
  jbrowse add-assembly $fastafiles --out . --load inPlace
  done
```

```bash
conda activate genometools
```

Lets use another loop to convert our gff files into sorted.gff files 

```bash
for gfffiles in *.gff
  do
  name=$(basename -s ".gff" $gfffiles)
  gt gff3 -sortlines -tidy -retainids $gfffiles > $name\.sorted.gff
  done
```

Now lets make another loop to bgzip our sorted files

```bash
for sortedfiles in *.sorted.gff
  do
  bgzip $sortedfiles
  done
```

lets make another loop to tabix our sorted.gff.g files

```bash
for gzfiles in *.sorted.gff.gz
  do
  tabix $gzfiles
  done
```

Finally lets make a loop that adds the track for all our sequences into jbrowse

```bash
for sortedgff in *.sorted.gff.gz
  do 
  name=$(basename -s ".sorted.gff.gz" $sortedgff)
  jbrowse add-track $sortedgff -a $name --out . --load inPlace  
  done
```

#### Now lets use the command to start up jbrowse and check out our sequences! 

```bash
npx serve .
```
